Решение основано на самообъединении таблицы по полям name и category, что формирует все возможные комбинации записей с одинаковыми характеристиками, а последующая фильтрация по условию id1 < id2 оставляет только уникальные несимметричные пары, исключая дубликаты  и записи с равными ID

```python
import pandas as pd
```

```python

items = pd.DataFrame({
    'id': [1, 2, 3, 4, 5],
    'name': ['A', 'B', 'A', 'C', 'B'],
    'category': ['X', 'Y', 'X', 'Z', 'Y']
})

merged = items.merge(items, on=['name', 'category'], suffixes=('_1', '_2'))

pairs = merged[merged['id_1'] < merged['id_2']]

result = pairs[['id_1', 'id_2', 'category']].rename(columns={'id_1': 'id1', 'id_2': 'id2'})

result
```
